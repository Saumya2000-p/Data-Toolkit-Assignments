```python
Answer3. Explain the following terms:
* The difference in np.array, np.asarray and np.asanyarray. 
* The difference between Deep copy and shallow copy.
```

```python
1. Differences among np.array, np.asarray, and np.asanyarray
np.array
* Usage: Creates a new array.
* Behavior: Always returns a new NumPy array regardless of the input type. It can take a list, tuple, or another array as input. It also has options for specifying the data type (dtype) and whether to copy the data.
```

```python
import numpy as np

a = np.array([1, 2, 3])
```

```python
np.asarray
* Usage: Converts input to an array, but does not copy if the input is already an array.
* Behavior: Returns an array if the input is an array or if it can be converted to an array. If the input is already a NumPy array, it will not make a copy, which can be more memory efficient.
```

```python
b = np.asarray(a)  # No copy if `a` is already an array
```

```python
np.asanyarray
* Usage: Similar to np.asarray, but allows for subclasses of arrays (like masked arrays).
* Behavior: Returns an array of any type (including subclasses). This is useful if you want to maintain the specific type of an input array.
```

```python
c = np.asanyarray(np.ma.array([1, 2, 3]))  # Preserves the masked array type
```

```python
Summary:-
* np.array: Always creates a new array (copy).
* np.asarray: Converts input to an array, does not copy if the input is already an array.
* np.asanyarray: Converts input to an array but allows subclasses (like masked arrays) to be returned as-is.
```

```python
2. Differences between Deep Copy and Shallow Copy
Shallow Copy
* Definition: A shallow copy creates a new object, but it inserts references into it to the objects found in the original. This means that changes to mutable objects in the original will reflect in the shallow copy.
```

```python
import copy

original = [[1, 2, 3], [4, 5, 6]]
shallow = copy.copy(original)

shallow[0][0] = 'X'  # Modifies the original as well
print(original)  # Output: [['X', 2, 3], [4, 5, 6]]
```

```python
Deep Copy
* Definition: A deep copy creates a new object and recursively adds copies of nested objects found in the original. This means that changes to mutable objects in the original do not affect the deep copy.
```

```python
deep = copy.deepcopy(original)

deep[0][0] = 'Y'  # Modifies only the deep copy
print(original)  # Output: [['X', 2, 3], [4, 5, 6]]
```

```python
Summary:-
* Shallow Copy: Copies the object but references nested objects; changes to nested objects affect both copies.
* Deep Copy: Copies the object and all nested objects; changes to nested objects in one copy do not affect the other.

These concepts are important for managing memory and ensuring data integrity when working with complex data structures in programming.
```

```python
import numpy as np

# Create a 5x6 array with random integers between 1 and 10
random_array = np.random.randint(1, 11, size=(5, 6))

print("Random Array (5x6):")
print(random_array)

# a) Extract all even integers from the array
even_integers = random_array[random_array % 2 == 0]
print("\nEven Integers:")
print(even_integers)

# b) Extract all odd integers from the array
odd_integers = random_array[random_array % 2 != 0]
print("\nOdd Integers:")
print(odd_integers)
```

```python
To filter and select rows from the People_Dataset where the "Last Name" column contains the name "Duke", the "Gender" column is "Female", and the "Salary" is less than 85,000, we can use the following code:

Step-by-Step Code:
1. Import Pandas: Ensure you have the Pandas library imported.
2. Read the Dataset: Load the dataset using pd.read_csv().
3. Filter the DataFrame: Apply conditions to filter the rows.
4. Display the Result: Show the filtered DataFrame.
```

```python
import pandas as pd

# Load the dataset (adjust the file path as necessary)
df = pd.read_csv('data.csv')

# Filter the DataFrame based on the specified conditions
filtered_df = df[
    (df['Last Name'].str.contains('Duke', case=False, na=False)) &  # Last Name contains 'Duke'
    (df['Gender'].str.contains('Female', case=False, na=False)) &   # Gender is 'Female'
    (df['Salary'] < 85000)                                         # Salary less than 85000
]

# Display the filtered DataFrame
print("Filtered Rows:")
print(filtered_df)
```

```python
Explanation of the Code:
1. Reading the Dataset: Loads the dataset into a DataFrame.
2. Filtering Conditions:
* df['Last Name'].str.contains('Duke', case=False, na=False): Checks if "Last Name" contains "Duke", ignoring case and handling missing values.
* df['Gender'].str.contains('Female', case=False, na=False): Checks if "Gender" is "Female".
* (df['Salary'] < 85000): Checks if "Salary" is less than 85,000.
* The conditions are combined using the & operator.
3. Displaying the Result: Prints the filtered DataFrame.
```

```python
We can create a time-series dataset in a Pandas DataFrame and then plot Temperature and Humidity using Matplotlib with different y-axes.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
dates = pd.date_range(start='2023-01-01', periods=100)

# Create a random dataset
temperature = np.random.uniform(low=20, high=35, size=100)  # Temperature in °C
humidity = np.random.uniform(low=30, high=90, size=100)     # Humidity in %

# Create a DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

# Set the Date column as the index
df.set_index('Date', inplace=True)

# a) Plot Temperature and Humidity
fig, ax1 = plt.subplots()

# Plot Temperature
ax1.plot(df.index, df['Temperature'], color='red', label='Temperature', marker='o')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df.index, df['Humidity'], color='blue', label='Humidity', marker='x')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# b) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

```python
Explanation:
1. Date Range: We create a date range for 100 days starting from January 1, 2023.
2. Random Dataset: Random values for Temperature and Humidity are generated within specified ranges.
3. DataFrame: The DataFrame is created with 'Date', 'Temperature', and 'Humidity' columns.
4. Plotting:
* We plot Temperature on the primary y-axis (left) and customize its appearance.
* We create a secondary y-axis using twinx() for Humidity.
* Each y-axis is labeled with the appropriate colors.
5. Title: We set the title of the plot.
```

```python
Answer15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
a) Plot a histogram of the data with 30 bins.
b) Overlay a line plot representing the normal distribution's probability density function (PDF).
c) Label the x-axis as 'Value' and the y-axis as Frequency/Probability.
d) Set the title of the plot as Histogram with PDF Overlay.
```

```python
We can create a histogram of normally distributed data and overlay the probability density function (PDF) using Matplotlib and NumPy.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# b) Overlay the normal distribution's PDF
xmin, xmax = plt.xlim()  # Get the range for x-axis
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, loc=0, scale=1)  # Calculate PDF for a normal distribution
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

# c) Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
Explanation:
1. Normal Distribution: The np.random.normal function generates 1000 samples with a mean of 0 and a standard deviation of 1.
2. Histogram: The hist function plots the histogram with 30 bins, normalized to show probabilities.
3. PDF Overlay: The PDF is computed using scipy.stats.norm.pdf and plotted on the same graph.
4. Labels and Title: The axes are labeled, and a title is added for clarity.
```

```python
Answer17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
We can create a scatter plot using Seaborn, where the color of the points is determined by their position relative to the origin, effectively dividing the plot into quadrants.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame to hold the data
data = pd.DataFrame({'X': x, 'Y': y})

# Define a function to categorize points based on their quadrant
def assign_quadrant(row):
    if row['X'] >= 0 and row['Y'] >= 0:
        return 'Quadrant 1'
    elif row['X'] < 0 and row['Y'] >= 0:
        return 'Quadrant 2'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

# Apply the function to create a new column in the DataFrame
data['Quadrant'] = data.apply(assign_quadrant, axis=1)

# Create a scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='Set1', s=100)

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title of the plot
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrants')

# Show the plot
plt.axhline(0, color='black',linewidth=0.8, ls='--')
plt.axvline(0, color='black',linewidth=0.8, ls='--')
plt.grid(True)
plt.show()
```

```python
Explanation:
1. Random Arrays: We generate two random arrays, x and y, with 100 samples each.
2. DataFrame: A pandas DataFrame is created to hold the x and y values.
3. Quadrant Assignment: A function determines the quadrant for each point based on its x and y coordinates, and this information is added as a new column in the DataFrame.
4. Seaborn Scatter Plot: We use Seaborn's scatterplot function to create the scatter plot, coloring points based on their quadrant.
5. Axes and Title: The axes are labeled, and the title is set.
6. Legend: A legend is added to identify the quadrants.
7. Grid and Axes: Dashed lines are drawn at the x and y axes to better visualize the quadrants.
```

```python
Answer18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.
```

```python
To create a line chart of a sine wave using Bokeh, we can follow these steps. Below is an example of how to do this, including grid lines, axis labels, and a title.
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable output in Jupyter Notebook (optional)
output_notebook()

# Create data for the sine wave
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2π
y = np.sin(x)                        # Sine values

# Create a new plot with title and axis labels
p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label='sin(x)', line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.3

# Show the plot
show(p)
```

```python
Explanation:
1. Data Creation: We use NumPy to generate x values from 0 to 2π and compute the corresponding sine values.
2. Bokeh Figure: We create a new Bokeh figure and set the title and axis labels.
3. Line Renderer: We add a line to the plot using the line method, specifying the x and y values.
4. Grid Lines: We adjust the transparency of grid lines with grid_line_alpha.
5. Display: Finally, we display the plot using show().
```

```python
Answer19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.
```

```python
We can create a bar chart of randomly generated categorical data using Bokeh, with colored bars based on their values and hover tooltips to display the exact values.
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Enable output in Jupyter Notebook (optional)
output_notebook()

# Generate random categorical data
categories = [f'Category {i}' for i in range(1, 11)]
values = np.random.randint(1, 100, size=10)  # Random values for each category

# Create a DataFrame for better handling
data = pd.DataFrame({'Category': categories, 'Value': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a new plot
p = figure(x_range=categories, title='Random Categorical Bar Chart', 
           toolbar_location=None, tools="")

# Create bars and color them based on values
colors = [f'#{int(val*255/100):02x}00{int(255-(val*255/100)):02x}' for val in values]  # Color gradient
p.vbar(x='Category', top='Value', source=source, width=0.9, color=colors)

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Show grid lines
p.grid.grid_line_color = "grey"
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)
```

```python
Explanation:
1. Data Generation: We create a list of categories and generate random values for each category.
2. Data Handling: A pandas DataFrame is created for easier manipulation, and then a ColumnDataSource is set up for Bokeh.
3. Bar Chart: We use the vbar function to create vertical bars, coloring them based on their values to create a gradient effect.
4. Hover Tooltips: We add a hover tool that displays the category and its corresponding value when the user hovers over a bar.
5. Axis Labels and Grid: We label the axes and enable grid lines for better visualization.
6. Display: Finally, the plot is displayed using show().
```

```python
Answer20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.
```

```python
We can create a basic line plot using Plotly with a randomly generated dataset by following these steps. Below is an example code snippet that demonstrates how to do this:
```

```python
import numpy as np
import plotly.graph_objects as go

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)  # X values
y = np.random.rand(100)  # Random Y values

# Create a line plot
fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Label the axes
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show the plot
fig.show()
```

```python
Explanation:
1. Data Generation: We create an array of x values ranging from 0 to 10, and generate corresponding random y values.
2. Plot Creation: We use Plotly's go.Figure() to create a new figure and add a trace for the line plot using go.Scatter().
3. Axes Labels and Title: The update_layout() method is used to set the title and label the axes.
4 Display: Finally, we display the plot using fig.show().
```

```python
Answer21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.
```

```python
We can create an interactive pie chart using Plotly with randomly generated data by following the code below. This example demonstrates how to add labels, percentages, and set the title for the pie chart:
```

```python
import numpy as np
import plotly.graph_objects as go

# Generate random data
np.random.seed(0)  # For reproducibility
labels = [f'Category {i}' for i in range(1, 6)]
values = np.random.randint(10, 100, size=5)  # Random values for each category

# Create an interactive pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values, 
                               textinfo='label+percent', 
                               hoverinfo='label+percent')])

# Set the title
fig.update_layout(title='Interactive Pie Chart')

# Show the plot
fig.show()
```

```python
Explanation:
1. Data Generation: We create a list of labels for the categories and generate random integer values for each category.
2. Pie Chart Creation: We create a pie chart using go.Pie(), where textinfo is set to display both labels and percentages, and hoverinfo shows the same information on hover.
3. Title: The title of the pie chart is set using update_layout().
4. Display: Finally, we display the interactive pie chart using fig.show().
```
